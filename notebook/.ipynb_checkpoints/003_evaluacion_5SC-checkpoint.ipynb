```python
import pandas as pd
from iertools.read import read_sql
import plotly.express as px
```

```python
def compara_ep_eh(fep,columna_eh):
    ep = read_sql(fep, alias=True).data
    ep = ep.asfreq("10min")
    # ep.reset_index(inplace=True)
    f = '../data/enerhabitat-2025-10-20.csv'
    eh = pd.read_csv(f,
                     index_col=0,
                     parse_dates=True,
                     usecols = ['Time', 'Tsa_1', 'Ti_1',"Ti_2","Ti_3","Ti_4","Ti_5"])
    eh.index= eh.index.tz_localize(None)
    # eh.reset_index(inplace=True)
    data = pd.concat([eh,ep],axis=1)
    data.reset_index(inplace=True)
    # print(data)
    fig = px.scatter(
        data_frame=data,
        x="index",
        y=[
            columna_eh,
            "Tsa_1",
            "Ti_CUBO"
        ],
        labels={"x": "Tiempo", "y": "Valor"}  # t√≠tulos de ejes
    )
    fig.show()
    
```

```python
compara_ep_eh("../idf/10cmAdobe.sql","Ti_1")
```

```python
compara_ep_eh("../idf/10cmLadrillo.sql","Ti_2")
```
